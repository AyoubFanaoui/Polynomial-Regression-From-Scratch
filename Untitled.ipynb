```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 6*np.random.rand(200, 1) 
y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200,1) # Ajout d'un bruit gaussien
```

```python
plt.plot(X,y, "b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

###  Construction du modèle de régression polynomiale "from scratch"
L'objectif est d'ajuster un modèle polynomial de degré donné en utilisant la régression linéaire sur des caractéristiques étendues.


### Transformation des données en polynômes

```python
def polynomial_features(X, degree):
    """
    Transforme X en une matrice contenant les puissances de X jusqu'au degré spécifié.
    """
    X_poly = np.ones((X.shape[0], degree + 1))  # Création d'une matrice remplie de 1 (pour inclure le biais)
    for i in range(1, degree + 1):
        X_poly[:, i] = X[:, 0]**i  # Ajout des puissances de X
    return X_poly
```

## Calcul des coefficients via la régression linéaire
Nous utilisons la formule des moindres carrés :




def fit_polynomial_regression(X, y, degree):
    """
    Ajuste un modèle de régression polynomiale en calculant les coefficients.
    """
    X_poly = polynomial_features(X, degree)  # Transformation de X en caractéristiques polynomiales
    theta = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ y  # Calcul des coefficients
    return theta


### Prédiction avec le modèle ajusté

```python
def predict(X, theta, degree):
    """
    Fait une prédiction en utilisant le modèle appris.
    """
    X_poly = polynomial_features(X, degree)
    return X_poly @ theta  # Produit matriciel pour obtenir les prédictions
```

### *Application du modèle et visualisation des résultats*

```python
degree = 10 # Choix du degré du polynôme
theta = fit_polynomial_regression(X, y, degree)  # Ajustement du modèle
```

### Prédiction sur des données triées pour une courbe lisse

```python
X_test = np.linspace(0, 6, 100).reshape(-1, 1)  # Génération de valeurs test
y_pred = predict(X_test, theta, degree)  # Prédictions

plt.scatter(X, y, color='blue', label="Données réelles")  # Données originales
plt.plot(X_test, y_pred, color='red', label="Régression polynomiale")  # Courbe de régression
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

## coefficient de détermination

```python
def r2_score(y_real, y_pred):
    """
    Calcule le coefficient de détermination R^2.
    """
    ss_total = np.sum((y_real - np.mean(y_real)) ** 2)  # Somme des carrés totaux
    ss_residual = np.sum((y_real - y_pred) ** 2)  # Somme des carrés résiduels
    return 1 - (ss_residual / ss_total)
```

```python
y_train_pred = predict(X, theta, degree)  # Prédictions sur les données d'entraînement
r2 = r2_score(y, y_train_pred)

print(f"Score R^2 du modèle : {r2 * 100:.2f}%")
```
